```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('package')
```

```python
x= df.iloc[:,0:1]
y= df.iloc[:, -1]   # x means input and y means output
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression #applied algorithm
```

```python
lr= LinearRegression()   #mDE OBJECTS
```

```python
lr.fit(x_train,y_train) #IDHR MODEL TRAIN HOGYA BHAIII
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))  #here ALGORITHM WORKS BECOZ JIS BNDE KA CGPA 8.58 THA USS BNDE KA PACKAGE 4 HAI which is same as test data
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(x_test, lr.predict(x_test), color='yellow')
plt.xlabel('CGPA')
plt.ylabel('package')
```
